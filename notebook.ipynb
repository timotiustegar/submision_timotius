## Visualization & Explanatory Analysis

### Pertanyaan 1: Apakah terdapat perbedaan pola polusi udara untuk setiap musim yang berbeda?

#### Fungsi untuk menentukan musim

```python
# Fungsi menentukan jenis musim berdasarkan bulan
def season_cond(month):
    if month in [3, 4, 5]:
        return 'Semi'
    elif month in [6, 7, 8]:
        return 'Panas'
    elif month in [9, 10, 11]:
        return 'Gugur'
    else:
        return 'Dingin'
```

#### Analisa pola polusi masing masing musim pada Aotizhongxin_df

```python
# Copy tabel sebelum melakukan analisa
aotizhongxin_season = aotizhongxin_df.copy()
# Satukan 3 feature waktu menjadi 1
aotizhongxin_season['date'] = pd.to_datetime(aotizhongxin_season[['year', 'month', 'day']])
# Group tabel perharinya
groupped_date = aotizhongxin_season.groupby("date").agg({
    "PM2.5": "mean",
    "PM10": "mean"
}).reset_index()
# Berikan klasifikasi label musim berdasarkan bulan sebagai ketentuannya
groupped_date["season"] = groupped_date['date'].dt.month.apply(lambda x : season_cond(x))

# Kelola subplot visualisasi
fig = plt.figure(figsize=[14,7])
grid = plt.GridSpec(nrows=2, ncols=4)
ax1 = plt.subplot(grid[0,0:4])
ax2 = plt.subplot(grid[1,0])
ax3 = plt.subplot(grid[1,1])
ax4 = plt.subplot(grid[1,2])
ax5 = plt.subplot(grid[1,3])

# Plot semua visualisasi yang dibutuhkan
sns.scatterplot(data=groupped_date, x="PM2.5", y="PM10", hue="season", palette={"Semi":"#da4167", "Panas":"#74a57f", "Gugur":"#ffd166", "Dingin":"#083d77"}, ax=ax1)
sns.scatterplot(data=groupped_date[groupped_date["season"]=="Semi"], x="PM2.5", y="PM10", color="#da4167", ax=ax2)
sns.scatterplot(data=groupped_date[groupped_date["season"]=="Panas"], x="PM2.5", y="PM10", color="#74a57f", ax=ax3)
sns.scatterplot(data=groupped_date[groupped_date["season"]=="Gugur"], x="PM2.5", y="PM10", color="#ffd166", ax=ax4)
sns.scatterplot(data=groupped_date[groupped_date["season"]=="Dingin"], x="PM2.5", y="PM10", color="#083d77", ax=ax5)
```

### Pertanyaan 2: Kapan waktu terakhir terjadi peningkatan tingkat polusi udara secara signifikan pada setiap stasiun?

```python
all_recency = all_df.copy()
all_recency['date'] = pd.to_datetime(all_recency[['year', 'month', 'day']])
station_list = all_recency['station'].unique()
for station in station_list:
    # Ambil data untuk masing masing stasiun
    station_data = all_recency[all_recency['station'] == station]
    
    # Filter semua data yang mengalami peningkatan polusi diatas 100 poin PM2.5 dibanding data sebelumnya
    significant_increase = station_data[(station_data['PM2.5'].diff() > 100) & (station_data['PM2.5'].shift(1)!=3)]

    # Mengambil data peningkatan polusi signifikan terakhir dan 1 data sebelum peningkatan polusi
    index_selected = station_data[station_data['No'] == significant_increase.tail(1)["No"].values[0]].index[0]
    selected_data = station_data.iloc[index_selected - 1 : index_selected + 1]

    if not significant_increase.empty:
        last_increase_date = significant_increase['date'].max()
        print(f"Terakhir kali terjadi peningkatan polusi di stasiun {station}: {last_increase_date.date()}")
        display(selected_data)
    else:
        print(f"Tidak ada peningkatan polusi di stasiun {station}")
```

## Conclusion

- Conclution pertanyaan 1 
Ketiga stasiun menunjukan pola yang hampir serupa pada keempat musim dimana ketiga statsiun pengukuran tidak menunjukan perbedaan pola polusi udara yang signifikan untuk masing masing musim. Namun berhasil ditemukan beberapa karakteristik perbedaan untuk setiap musimnya seperti, pada musim semi partikel polusi yang berukuran lebih besar yaitu sampai pada 10mm lebih banyak ditemukan dibandingkan dengan musim lain, musim panas menjadi musim dengan tingkat polusi terendah sedangkan musim dingin selalu menjadi musim dengan tingkat polusi tertinggi untuk seluruh stasiun pengukuran. 
Dengan mengetahui fakta ini, pemerintah kota masing masing pengukuran dapat melakukan beberapa macam hal seperti, memfokuskan filter udara untuk partikel yang lebih besar hanya pada musim semi dan mencoba mengidentifikasi penyebab selalu tingginya polusi udara pada musim dingin di hampir semua stasiun.
- Conclution pertanyaan 2
Stasiun Aotizhongxin dan Changping terakhir kali mengalami kenaikan polusi udara secara signifikan diatas 100 poin PM2.5 yaitu pada tanggal 28 Januari 2017, dimana stasiun Aotizhongxin mencatat kenaikan sebesar 271 pada pukul 1 dini hari sedangkan stasiun Changping mencatat kenaikan sebesar 238 pada pukul yang sama. Stasiun Guanyuan mencatat kenaikan signifikan terakhirnya berada pada bulan Februari tepatnya tanggal 11 Februari 2017 sebesar 123 pada pukul 10 malam.
Dengan mengetahui fakta ini, pemerintah kota dapat mencoba untuk mengidentifikasikan penyebab dari peristiwa kenaikan polusi udara yang cukup signifikan ini sehingga bisa mengatasi sekaligus mencegah kenaikan polusi terjadi kembali dengan alasan yang sama.
- Conclution Geoanalysis
Stasiun Changping yang lebih sepi penduduk dibandingkan dengan Aotizhongxin memiliki tingkat polusi udara yang lebih baik baik itu untuk partikel besar 10mm atau partikel kecil 2.5mm. Hal ini dapat dilihat dari persebaran data PM2.5 dan PM10 milik stasiun Chanping yang lebih condong ke kiri menunjukan bahwa lebih banyak waktu pemeriksaan yang mencatat tingkat polusi udara lebih sedikit. Selain itu, kandungan zat yang ada diudara pada stasiun Changping juga relatif lebih baik, dimana kadar Sulfur Dioksida, Nitrogen Dioksida, dan Karbon Monoksida dalam udara pada stasiun Changping lebih sedikit dibandingkan dengan stasiun Aotizhongxin dan Guanyuan.
Dengan mengetahui fakta ini, dapat dilihat sebagai sample bahwa kepadatan penduduk suatu kota dapat mempengaruhi kualitas udaranya dimana semakin padat penduduk akan lebih berpotensi memiliki tingkat polusi udara yang lebih tinggi. Kesimpulan ini diambil berdasarkan analisa yang menemukan bahwa Changping dengan penduduk yang lebih sedikit memiliki kualitas udara lebih baik dibandingkan dengan Aotizhongxin dan Guanyuan
